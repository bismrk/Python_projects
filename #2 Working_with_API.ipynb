```python
# обрізаєм колнку event_time для подальшого використання
df['event_time'] = df['event_time'].str[:10]
df['event_time']
```

```python
import requests
```

```python
df.currency.unique()
```

```python
len(df['event_time'].unique())
```

```python
eur_usd = {}
for time in df['event_time'].unique():
  responce = requests.get(f'https://api.frankfurter.app/{time}?from=EUR&to=USD')
  responce = responce.json()
  eur_usd[time] = responce['rates']['USD']
eur_usd
```

```python
df['eur_usd'] = df['event_time'].map(eur_usd)
df.head(6)
```
